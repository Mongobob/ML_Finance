```python
Y = pd.factorize(mapping["splticrm"])[0]
print(np.bincount(Y)) 
print(pd.factorize(mapping["splticrm"])[1])
Y = pd.DataFrame(Y, columns=["Rating as Factor"])
print(Y)
```

```python
X = mapping.loc[:, mapping.columns != "splticrm"]
X = X.drop(["permno", "CUSIP", "NCUSIP", "adate", "qdate", "public_date", "TICKER"], axis=1)
X = X.drop(["COMNAM", "PERMCO", "NWPERM", "gvkey", "datadate", "tic", "cusip", "conm", "PRC"], axis=1) #prc = price on stock market
X.head()
```

```python
NAs = X.isnull().sum() > 10000
Zeros =  (X == 0).sum() > 10000
delNAs = X.columns[NAs] #drops PEG_trailing
delZeros= X.columns[Zeros] #drops rd_sale, adv_sale, staff_sale
X = X.drop(delNAs, axis=1)
X = X.drop(delZeros, axis=1)
```

```python
#Fill NA's with ffill
X_ffill = X.fillna(axis="columns", method="ffill")
X_ffill["CAPEI"] = X_ffill["CAPEI"].fillna(method="ffill")
X_ffill.shape
```

```python
#Fill NA's with zero
X_fillZero = X.fillna(value=0)
X_fillZero.shape
```

```python
#Delete rows with any NA's
total = pd.concat([X, Y], axis=1) #concatenate X and Y
total = total.dropna() #drop all rows with NAs

X_delNA = total.iloc[:,:-1] #New X
Y_delNA = total.iloc[:,-1] #New Y

X_delNA.shape
```

```python
#Define different training-test sets of different methods for dealing with NAs
X_train_ffill = X_ffill.iloc[:25000,:]
X_test_ffill = X_ffill.iloc[25000:,:]

X_train_fillZero = X_fillZero.iloc[:25000,:]
X_test_fillZero = X_fillZero.iloc[:25000,:]

X_train_delNA = X_delNA.iloc[:10575,:]
X_test_delNA = X_delNA.iloc[:10575,:]

Y_train = Y.iloc[:25000]
Y_test = Y.iloc[25000:]

Y_train_delNA = Y.iloc[:10575]
Y_test_delNA = Y.iloc[10575:]
```

```python
from sklearn.ensemble import RandomForestClassifier

def feature_selection(x, y, n):
    """ Find out, which the most important features are.
    
    Args:
        x: feature input without NaN values
        y: classification input
        n: number of most important features, that need to be calculated
    """
    
    feat_labels = x.columns[:]
    
    # Create Random Forest object, fit data and
    # extract feature importance attributes
    forest = RandomForestClassifier(random_state=1, class_weight= "balanced")
    forest.fit(x, y)
    importances = forest.feature_importances_
    
    # Get cumsum of the n most important features
    feat_imp = np.sort(importances)[::-1]
    sum_feat_imp = np.cumsum(feat_imp)[:n]
    
     # Sort output (by relative importance) and 
    # print top n features
    indices = np.argsort(importances)[::-1]
    for i in range(n):
        print('{0:2d}) {1:7s} {2:6.4f}'.format(i + 1, 
                                           feat_labels[indices[i]],
                                           importances[indices[i]]))
    
    # Plot Feature Importance (both cumul., individual)
    plt.figure(figsize=(12, 8))
    plt.bar(range(n), importances[indices[:n]], align='center')
    plt.xticks(range(n), feat_labels[indices[:n]], rotation=90)
    plt.xlim([-1, n])
    plt.xlabel('Feature')
    plt.ylabel('Rel. Feature Importance')
    plt.step(range(n), sum_feat_imp, where='mid', 
         label='Cumulative importance')
    plt.tight_layout();
    
```

```python
feature_selection(x = X_train_ffill, y = Y_train.values.ravel(), n = 40) #.values will give the values in an array.(shape: (n,1)),  .ravel will convert that array shape to (n, )
```

```python
feature_selection(x = X_train_fillZero, y = Y_train.values.ravel(), n = 67)
```

```python
feature_selection(x = X_train_delNA, y = Y_train_delNA.values.ravel(), n = 67)
```
